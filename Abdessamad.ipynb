```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from google.colab import files
uploaded = files.upload()
```

```python
Dataframe =pd.read_csv("Abdessamad.csv", delimiter = ";")
Dataframe.head()
```

```python
features = Dataframe.iloc[:,:35]
labels = Dataframe.iloc[:,35:]
labels
```

```python
from sklearn.model_selection import train_test_split
train_features , test_features = train_test_split(features,test_size = 0.1, random_state=42)
train_labels , test_labels = train_test_split(labels,test_size = 0.1, random_state=42)
```

```python
#train_features.shape
```

```python
features.corr()
plt.figure(figsize=(26,20))
sns.heatmap(features.corr(), annot=True)
```

```python
Model = tf.keras.Sequential([
  layers.Dense(516, activation="relu"),
  layers.Dense(258, activation="relu"),
  layers.Dense(11, activation ="softmax")
])
Model.compile(loss =["mse"], optimizer = tf.optimizers.Adam(), metrics=["accuracy", "mae"])
history = Model.fit(train_features, train_labels, epochs=100)
#Recla_model.predict(features)
#Recla_model.summary()
```

```python
import matplotlib.image  as mpimg
import matplotlib.pyplot as plt

#-----------------------------------------------------------
# Retrieve a list of list results on training and test data
# sets for each training epoch
#-----------------------------------------------------------
mae=history.history['accuracy']
loss=history.history['loss']

epochs=range(len(loss))

#------------------------------------------------
# Plot MAE and Loss
#------------------------------------------------
plt.plot(epochs, mae, 'r')
plt.plot(epochs, loss, 'b')
plt.title('MAE and Loss')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["MAE", "Loss"])

plt.figure()

epochs_zoom = epochs[200:]
mae_zoom = mae[200:]
loss_zoom = loss[200:]
```
